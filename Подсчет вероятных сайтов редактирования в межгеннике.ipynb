```python
def chroms_list() -> list:
    """Функция создает список реальных хромосом."""
    chroms = ['chr'+ str(i) for i in range(1,23)]
    chroms.append('chrX')
    chroms.append('chrY')
    return chroms
```

```python
def intergene_search(chrom: str, start: int, end: int) -> str:
    """Функция принимает на вход хромосому и координаты в ней. Выводит подстроку генома"""
    with pysam.FastaFile(f'{path}/GRCh38_24.fasta') as fh:
        gene_str = fh.fetch(chrom, start, end)
    return gene_str
```

```python
def teor_esites_position(chrom: str) -> list:
    """Функция на вход принимает имя хромосомы. Возвращает список с вакантными координатами мест редактирования в межгеннике"""
    chr_list = []
    if chrom == 'chr1':
        for j, intergene in enumerate(intergenic[n[chrom][0]: n[chrom][1]]):#при срезе бед файла берется [1;10) (невключительно справа). Но тк структура n учитывает это, то все нормально.Но так как структура n имеет свою специфику (смотреть описание к n), то приходится разбивать функцию на 2 части. 
            intergene_by_letter = intergene_search(chrom, int(intergene[1]), int(intergene[2]))
            matching_letter = ['A','T']
            chr_sublist = (int(k[j][1])+i for i, letter in enumerate(intergene_by_letter) if letter in matching_letter)
            chr_list.extend(chr_sublist)
    else:
        for j, intergene in enumerate(intergenic[n[chrom][0]-1: n[chrom][1]]):#при срезе бед файла берется [1;10) (невключительно справа). Но тк структура n учитывает это, то все нормально.
            intergene_by_letter = intergene_search(chrom, int(intergene[1]), int(intergene[2]))
            matching_letter = ['A','T']
            chr_sublist = (int(k[j][1])+i for i, letter in enumerate(intergene_by_letter) if letter in matching_letter)
            chr_list.extend(chr_sublist)
    chr_list = np.array(chr_list, dtype = np.int32)
    print(chrom)
    return chr_list
```

```python
def teor_esites_position(chrom):
    chr_list = []
    for j, intergene in enumerate(intergenic[n[chrom][0]: n[chrom][1]]):
        intergene_by_letter = intergene_search(chrom, int(intergene[1]), int(intergene[2]))
        matching_letter = ['A','T']
        chr_sublist = (int(k[j][1])+i for i, letter in enumerate(intergene_by_letter) if letter in matching_letter)
        chr_list.extend(chr_sublist)
    chr_list = np.array(chr_list, dtype = np.int32)
    print(chrom)
    return chr_list
```

```python
pool = mp.Pool(mp.cpu_count())
results = pool.map(teor_esites_position, {chrom: [] for chrom in n.keys()})
pool.close()
```

```python
result_intergene_with_chrom_names = {chrom: [] for chrom in n.keys()}
for i, j in zip(result_intergene_with_chrom_names, results):
    result_intergene_with_chrom_names[i] = j
```
