```python
error_GTEX = ['GTEX-11ZUS-0011-R10b-SM-5BC79', 'GTEX-OHPN-0011-R10A-SM-5S2U3', 'GTEX-T6MN-0011-R10A-SM-5S2SB', 'GTEX-RVPU-0011-R10A-SM-2XCAH', 'GTEX-RU72-0011-R10A-SM-2TF6D', 'GTEX-12WSD-0011-R10b-SM-5GU79', 'GTEX-12WS9-0011-R10a-SM-57WBV', 'GTEX-T5JC-0011-R10A-SM-5S2TK', 'GTEX-OHPN-0011-R10A-SM-2I5FL']
list_of_columns_names_on_dtypes_for_BFC_BA9 = list(dict_of_columns_names_on_dtypes_for_BFC_BA9.keys())
for GTEX in error_GTEX:
    list_of_columns_names_on_dtypes_for_BFC_BA9.remove(GTEX)
list_of_columns_names_on_dtypes_for_BFC_BA9.insert(0, 'Description')
list_of_columns_names_on_dtypes_for_BFC_BA9.insert(0, 'Name')
```

```python
len(list_of_columns_names_on_dtypes_for_BFC_BA9)
```

```python
rna_seq_data = pd.read_csv(f'{path}/GTEx_Analysis_2017-06-05_v8_RNASeQCv1.1.9_gene_tpm.gct', skiprows = 2, 
                           sep = '\t', chunksize = 100, 
                           dtype = dict_of_columns_names_on_dtypes_for_BFC_BA9,
                          usecols = list_of_columns_names_on_dtypes_for_BFC_BA9)#вывести в том порядке, как нам надо, не получилось. Поэтому сделаем это постфактум
df_expressions = pd.concat(rna_seq_data)
```

```python
df_expressions = df_expressions[list_of_columns_names_on_dtypes_for_BFC_BA9]
df_expressions
```

```python
new_final_mapping_for_BFC_BA9 = dict()
for sample in final_mapping_for_BFC_BA9:
    if final_mapping_for_BFC_BA9[sample] in error_GTEX:
        continue
    else:
        new_final_mapping_for_BFC_BA9[sample] = final_mapping_for_BFC_BA9[sample]
new_final_mapping_for_BFC_BA9
```

```python
df_AEI_and_ADAR_expr_by_SRR_BFC_BA9 = df_AEI_and_ADAR_expr_by_SRR_BFC_BA9.query(
    f"Sample in {list(new_final_mapping_for_BFC_BA9.keys())}").reset_index(drop=True)
```

```python
AEI_row = list(df_AEI_and_ADAR_expr_by_SRR_BFC_BA9.AEI)
AEI_row
```

```python
len(AEI_row)
```

```python
AEI_row.insert(0, 'AEI')
AEI_row.insert(1, 'Alu-index')
```

```python
AEI_row
```

```python
df_expressions_copy = df_expressions.copy()
df_expressions_copy.loc[len(df_expressions_copy.index)] = AEI_row
df_expressions_copy
```

```python
df_expressions_copy.query("Description == 'ADAR'")
```

```python
dict_of_spear_kor = dict()
for gene in df_expressions.Name:
    gene_expression = df_expressions_copy.query(f"Name == '{gene}'").reset_index(drop = True).loc[0][2:].to_list()
    k_sp, p_value = stats.spearmanr(AEI_row[2:], gene_expression, alternative = 'two-sided')
    dict_of_spear_kor[gene] = (k_sp, p_value)
```

```python
dict_of_spear_kor
```

```python
list_of_k_sp = list()
list_of_p_value = list()
for gene, speatmen_result in dict_of_spear_kor.items():
    list_of_k_sp.append(speatmen_result[0])
    list_of_p_value.append(speatmen_result[1])
```

```python
df_spearmen = pd.DataFrame(data = {'gene': dict_of_spear_kor.keys(),'k_sp': list_of_k_sp, 'p_value': list_of_p_value})
df_spearmen
```

```python
df_spearmen.dropna(inplace = True)
df_spearmen
```

```python
pval_adj = multipletests(df_spearmen.p_value, alpha=0.01, method='fdr_bh', is_sorted=False, returnsorted=False)
pval_adj
```

```python
df_spearmen['pval_adj'] = pval_adj[1]
df_spearmen
```

```python
df_spearmen.to_csv(f'{path}/spearmen_correlation_for_all_genes_with_AEI_for_Frontal Cortex_(BA9).csv.gz', sep='\t',
                 index=False,
                 compression='gzip'
                 )
```

```python
very_cool_predictors = df_spearmen.query("(pval_adj < 0.001) and (k_sp > 0.70)").reset_index(drop=True)
very_cool_predictors
```
